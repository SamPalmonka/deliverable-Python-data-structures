```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris = iris.drop(labels = "species", axis = 1)
iris.head()
```

```python
iris["sepal_sum"] = iris["sepal_length"] + iris["sepal_width"]
iris.head()
```

```python
iris["petal_sum"] = iris["petal_length"] + iris["petal_width"]
iris.head()
```

```python
mean = (iris.mean())
std = (iris.std())
min = (iris.min())
max = (iris.max())

summary_dict = {"mean" :mean , "std" :std , "min" :min , "max" :max }
summary_df = pd.DataFrame(data = summary_dict)
summary_df
```
